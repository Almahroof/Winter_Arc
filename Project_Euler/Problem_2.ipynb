Problem 2: Sum of Even Fibonacci Numbers
Find the sum of all the even-valued terms in the Fibonacci sequence that do not exceed four million.

Problem Restatement
The Fibonacci sequence is defined by the recurrence relation:

$$
F_{n} =F_{n-1}+F_{n-2}
$$,
starting with $F_{1} = 1$ and $F_{1} = 1$.However, in this problem, we start with $F_{1} = 1$ and $F_{2} = 2$. The task is to calculate the sum of all even-valued Fibonacci numbers below a given limit, in this case, four million.

Direct Approach
A direct translation of the problem statement into code would look like this:

```python
# Initialize variables
limit = 4000000
sum = 0
a = 1
b = 1

# Loop to generate Fibonacci sequence and sum even terms
while b < limit:
    if b % 2 == 0:
        # sum= sum+b
        sum += b
    h = a + b
    a = b
    b = h

# Output the result
sum
```
